```python
# 예제 5-51 라이브러리 임포트
import pandas as pd
import folium
import json
```

```python
# 예제 5-53 서울시 스타벅스 매장 목록 데이터 불러오기
seoul_starbucks = pd.read_excel('./files/4_4_seoul_starbucks_list.xlsx')
seoul_starbucks.head()
```

```python
# 예제 5-54 folium을 이용한 지도 생성
starbucks_map = folium.Map(
    location=[37.573050, 126.979189],
#    tiles='Stamen Terrain',Stamen Toner,Stamen Watercolor, Mapbox Bright 적용안되고 일반 지도 스타일 OpenStreetMap이것만 됨.
#    OpenStreetMap은 기본 내장되어있기 때문.
#    배경화면 바꾸고 싶으면 참조: https://wooiljeong.github.io/python/folium-google-tiles/    
    
    zoom_start=11
)
starbucks_map
```

```python
# 예제 5-55 지도에 스타벅스 매장 위치를 나타내는 서클 마커 그리기
for idx in seoul_starbucks.index:
    lat = seoul_starbucks.loc[idx, '위도']
    lng = seoul_starbucks.loc[idx, '경도']

    folium.CircleMarker(
        location=[lat, lng],
        fill = True, 
        fill_color='green', 
        fill_opacity=1,
        color='yellow', 
        weight=1,
        radius=3
    ).add_to(starbucks_map)

starbucks_map.save("2.html")
```

```python
# 예제 5-56 스타벅스 매장 타입별 위치 서클 마커 그리기
starbucks_map2 = folium.Map(
    location=[37.573050, 126.979189],
    zoom_start=11
)

for idx in seoul_starbucks.index:
    lat = seoul_starbucks.loc[idx, '위도']
    lng = seoul_starbucks.loc[idx, '경도']
    store_type = seoul_starbucks.loc[idx, '매장타입']
    
    # 매장 타입별 색상 선택을 위한 조건문
    fillColor = ''
    if store_type == 'general':
        fillColor = 'gray'
        size = 1
    elif store_type == 'reserve':
        fillColor = 'blue'
        size = 5
    elif store_type == 'generalDT':
        fillColor = 'red'
        size = 5

    folium.CircleMarker(
        location=[lat, lng],
        color=fillColor,
        fill = True,
        fill_color = fillColor, 
        fill_opacity = 1,
        weight = 1,
        radius = size
    ).add_to(starbucks_map2)

starbucks_map2
```
